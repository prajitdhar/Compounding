```python
contexts=pd.read_pickle('/data/dharp/compounds/datasets/contexts/contexts_top50k.pkl')
```

```python
div_lsts=np.array_split(pkl_files, 10)
```

```python
len(div_lsts[0])
```

```python
def mem_reducer(pkl_file):
    print(f'\nStarted with file {pkl_file}\n')
    cur_time=time.time()
    
    df=pd.read_pickle(pkl_file)
    df=df.loc[df.context.isin(contexts)]
    
    df=df.groupby(['modifier','head','year'])['count'].sum().to_frame().reset_index()
    df=df.loc[df['year']>=1800]

    
    print(f'Done with file {pkl_file} in {round(time.time()-cur_time)} secs')
    return df
```

```python
entire_context_df_list=[]
for i,cur_list in enumerate(div_lsts):
    print(f'List num {i+1}')
    n_proc = len(cur_list)

    pool = Pool(n_proc)
    results=pool.map_async(mem_reducer,cur_list) 
    pool.close()
    pool.join()
    
    dfs=results.get()
    
    cur_df=pd.concat(dfs,ignore_index=True,sort=True)
    #cur_df=mem_reducer(pkl_file)

    display(cur_df.info())
    entire_context_df_list.append(cur_df)
```

```python
import pickle
```

```python
compounds_df=pd.concat(entire_context_df_list,ignore_index=True,sort=True)
```

```python
compounds_df=compounds_df.groupby(['modifier','head','year'])['count'].sum().to_frame().reset_index()
```
