```python
import pandas as pd
import glob
```

```python
words = pd.concat([pd.read_csv(f,sep="\t") for f in glob.glob('/data/dharp/compounding/datasets/words_*.csv')], ignore_index = True)
```

```python
words=words.groupby(['word','context','decade']).sum()
```

```python
words.reset_index(inplace=True)
```

```python
words.to_csv('/data/dharp/compounding/datasets/words.csv',index=False)
```
