# Data preprocessing


In order to fit the datasets in the neural network, two separate datasets are created - **modifier_constituents** and **head_constituents** .

For both datasets only those constituents are selected that have some statistics for being either a head or a modifier or both.

```python
constituents=pd.read_csv('constituents.csv',encoding='windows-1252',index_col=[0])
#constituents['pattern']=constituents['pattern'].astype('category')
#constituents['presence']=constituents['presence'].astype('category')
#constituents['presence_as_modifier']=constituents['presence_as_modifier'].astype('category')
#constituents['presence_as_head']=constituents['presence_as_head'].astype('category')
constituents.drop('pattern',axis=1,inplace=True)
constituents['presence']=1+np.log10(constituents['presence'])
constituents['presence_as_modifier']=1+np.log10(constituents['presence_as_modifier'])
constituents['presence_as_head']=1+np.log10(constituents['presence_as_head'])
constituents.head(20)
```
